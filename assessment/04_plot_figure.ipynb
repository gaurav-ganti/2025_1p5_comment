```python
import pyam
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
from matplotlib.patches import Rectangle
import matplotlib.transforms as mtransforms
import seaborn as sns

import numpy as np
```

```python
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 12
plt.rcParams["svg.fonttype"] = "none"
```

```python
df_sr15 = pyam.IamDataFrame('data/01_sr15_data.xlsx')
df_ar6 = pyam.IamDataFrame('data/01_ar6_data.xlsx')
ngfs = pyam.IamDataFrame('data/01_ngfs_data.xlsx')
debt = pyam.IamDataFrame('data/02_data.xlsx')
ghg = pd.read_excel('data/03_data.xlsx')
```

```python
def plot_range_max_eoc(df, qs, axs, loc, color):
    # First pull out max values
    m = df.timeseries().max(1)
    median = np.median(m)
    axs.fill_between(
        x=[loc-0.04, loc+0.04],
        y1=m.quantile(q=qs[0]),
        y2=m.quantile(q=qs[1]),
        color=color,
        alpha=0.6
    )
    axs.scatter(
        x=loc,
        y=median,
        color=color,
        marker='_',
        s=500
    )
```

```python
fig, ax = plt.subplot_mosaic(
    [
        ['a', 'a', 'a', 'a', 'a'],
        ['b', 'b', 'b', 'b', 'c'],
    ],
    figsize=(7,12),
)

# Panel a: Plot the range of peak warming outcomes
plot_range_max_eoc(df_sr15.filter(category='Below 1.5C'), [0.05,0.95], ax['a'], 0, '#1E638A')
plot_range_max_eoc(df_sr15.filter(category='1.5C low overshoot'), [0.05,0.95], ax['a'], 0.5, '#1E638A')
plot_range_max_eoc(df_ar6.filter(Category='C1'), [0.05,0.95], ax['a'], 1, '#F48128')
plot_range_max_eoc(ngfs, [0.05,0.95], ax['a'], 1.5, '#C3161C')
ax2 = ax['a'].twiny()
ax['a'].set_xlim([-0.5, 2])
ax2.set_xlim([-0.5, 2])
ax2.set_xticks([0, 0.5, 1, 1.5])
ax2.set_xticklabels(
    [2018, 2018, 2022, 2024],
    #weight='bold'
)
ax['a'].set_xticks([0, 0.5, 1, 1.5])
ax['a'].set_xticklabels(
    ['IPCC SR1.5\nBelow\n1.5°C', 'IPCC SR1.5\n1.5°C low\novershoot', 'IPCC AR6\nC1', 'NGFS\nPhase V']
)

#ax['a'].xaxis.tick_top()
ax['a'].spines[['top', 'bottom', 'right']].set_visible(False)
ax2.spines[['top', 'bottom', 'right']].set_visible(False)
ax['a'].grid(axis='y', alpha=0.2)
ax['a'].axhline(1.5, color='#31A18A', alpha=0.4)
ax['a'].axhline(1.6, color='#31A18A', alpha=0.4, linestyle='dashed')
ax['a'].axhline(1.7, color='#31A18A', alpha=0.4, linestyle='dotted')
ax2.tick_params(axis='both', which='both', length=0)
ax['a'].tick_params(axis='both', which='both', length=0)
ax['a'].set_ylabel('Peak warming with a 50% chance [°C]')

#ax['a'].plot(1, 3.5, ">k", transform=ax['a'].get_yaxis_transform(), clip_on=False)

## Panel B ###
scen_color_map = {
    'E':'#C3161C',
    'IMP-REN':'#1E638A',
}
for scen in scen_color_map.keys():
    debt.filter(
        scenario=scen,
        region='World',
        year=range(1990, 2051)
    ).plot(color=scen_color_map[scen], ax=ax['b'])
    debt.filter(
        scenario=scen,
        region='R10AFRICA',
        year=range(1990, 2051)
    ).plot(color=scen_color_map[scen], ax=ax['b'], linestyle='dotted')
    debt.filter(
        scenario=scen,
        region='R10EUROPE',
        year=range(1990, 2051)
    ).plot(color=scen_color_map[scen], ax=ax['b'], linestyle='dashed')

ax['b'].spines[['top', 'right', 'bottom']].set_visible(False)
ax['b'].axhline(100, color='#31A18A', alpha=0.4)
ax['b'].set_title('')
ax['b'].set_xlabel('')
ax['b'].grid(axis='y', alpha=0.2)
ax['b'].tick_params(axis='both', which='both', length=0)

custom_legend = [
    Line2D([0], [0], color='#C3161C', lw=2,  label=''),
    Line2D([0], [0], color='#1E638A', lw=2, label=''),
    Line2D([0], [0], color='#C3161C', linestyle='dashed', lw=2,  label=''),
    Line2D([0], [0], color='#1E638A', linestyle='dashed', lw=2, label=''),
    Line2D([0], [0], color='#C3161C', linestyle='dotted', lw=2),
    Line2D([0], [0], color='#1E638A', linestyle='dotted', lw=2),
    
]
ax['b'].legend(handles=custom_legend, framealpha=0, ncol=3)
ax['b'].set_ylabel('Consumed share of regional\ncarbon budget from 1990 [%]')
ax['c'].spines[['top', 'right', 'bottom', 'left']].set_visible(False)
ax['c'].set_xticks([])
ax['c'].set_yticks([])


## Subplot labels ##
for label, a in ax.items():
    if label in ['c', 'e']:
        continue
    if label=='d':
        label='c'
    trans = mtransforms.ScaledTranslation(-22/72, 10/72, fig.dpi_scale_trans)
    a.text(0.0, 1.0, label.upper(), transform=a.transAxes + trans, weight='bold', fontsize=10)
```

```python
fig.savefig('figure1_v3.svg', dpi=300, bbox_inches='tight')
fig.savefig('figure1_v3.png', dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(
    figsize=(6,7)
)

## Scatter the peak and 2100 warming
sns.scatterplot(
    data=ghg,
    x='warming_peak',
    y='warming_2100',
    ax=ax,
    hue='cumulative_netneg_gtco2',
    palette='icefire_r'
)

ax.annotate(
    '', 
    xy=(1.2, 1), 
    xytext=(1.15, 1.15), 
    arrowprops=dict(arrowstyle='->', color='grey', connectionstyle='arc3,rad=-0.3', alpha=0.5)
)

ax.annotate(
    'Scenarios cool\nafter peaking',
    xy=(1.33, 1.07),
    color='grey',
    alpha=.5,
    ha='center',
)

ax.axhline(1.5, color='#31A18A', alpha=0.5)
ax.axline((0,0), slope=1, color='grey', linestyle='dashed', alpha=0.5)

ax.set_xlim([1, 2.1])
ax.set_ylim([1, 2.1])
ax.get_legend().remove()

ax.spines[['top', 'right', 'bottom']].set_visible(False)
ax.set_title('')
ax.set_xlabel('')
ax.grid(axis='y', alpha=0.2)
ax.tick_params(axis='both', which='both', length=0)
ax.set_xticks([1.2, 1.4, 1.6, 1.8, 2.0])

ax.set_xlabel('Peak warming with a 50% chance [°C]')
ax.set_ylabel('2100 warming with a 50% chance [°C]')

# Making a custom colorbar for the third plot
norm = plt.Normalize(ghg['cumulative_netneg_gtco2'].min(), ghg['cumulative_netneg_gtco2'].max())
sm = plt.cm.ScalarMappable(cmap='icefire_r', norm=norm)
ax.figure.colorbar(sm, ax=ax, label='Cumulative net-negative GHG emissions [Gt CO$_2$e]', aspect=40, orientation='horizontal')
```
